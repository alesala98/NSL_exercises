The task of this exercise is to write a code in order to solve the Traveling Salesman Problem (TSP) using a genetic algorithm. The attached C++ code is able to solve this problem given any coherent number of cities to be visited and, in particular, it satisfies all the requirements of the provided text:

- the path of the salesman is expressed using a 1D representation while a check function verifies that any kind of path fulfills the necessary boundary conditions
- the population in each generation is sorted using $L^{(1)}$ loss/cost function, selecting specifically those individuals that minimize it
- the selected individuals of each generation are subject to mutation operations (pair-permutations, block-permutations, shifts and inversions which maintain the first city fixed in its position) and crossover operations, which are implemented following the steps described in the text.

Once the code was written, some simulations were performed in order to determine the best path (i.e. the path that minimizes $L^{(1)}$ function) between 32 cities. These cities were first generated along a circumference of radius $r=1$. For the simulations, it was decided to use $N_{generations} = 2000$, $N_{individuals} = 300$ (in each generation) and as the exponent of the rigged-roulette choice system for crossover operations $p = 2$. 

Below, you can find the results of these simulations. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Circular world
f, ax = plt.subplots(1,2)

gen, L = np.loadtxt("Datas/ave_L1_circ.dat", usecols=(0,1), unpack='true')
L_best = np.loadtxt("Datas/best_L1_circ.dat", usecols=1, unpack='true')
ax[0].plot(gen, L, label="Average")
ax[0].plot(gen,L_best, label="Minimum")
ax[0].set_title('$L^{(1)}$ (circumference)')
ax[0].set(xlabel="#Generations", ylabel='$ <L^{(1)}> $')
ax[0].legend()
ax[0].grid(True)

x, y = np.loadtxt("Datas/path_circ.dat", usecols=(0,1), unpack='true')
ax[1].plot(x, y, marker = "o")
ax[1].set_title("Best path (circumference)")
ax[1].set(xlabel="x", ylabel="y")
ax[1].grid(True)

plt.subplots_adjust(left=0.2, bottom=0., right=2.2, top=1.2, wspace=0.3, hspace=0.4) 

plt.show()
```

It is clear that for the previous choice of parameters, the algorithm is able to find the circumference itself as shortest path, which seems perfectly reasonable. At the same time, $\langle L^{(1)}\rangle$ averaged on the best half of the population tends to the minimum value as the number of generations increases.

Then, using the same parameters as above, the simulation is repeated this time creating an array of 32 cities randomly distributed inside a square of side $l = 1$. Below, you can find the results.

```python
#Square world
f, ax = plt.subplots(1,2)

gen, L = np.loadtxt("Datas/ave_L1_square.dat", usecols=(0,1), unpack='true')
L_best = np.loadtxt("Datas/best_L1_square.dat", usecols=1, unpack='true')
ax[0].plot(gen, L, label="Average")
ax[0].plot(gen,L_best, label="Minimum")
ax[0].set_title('$L^{(1)}$ (square)')
ax[0].set(xlabel="#Generations", ylabel='$ <L^{(1)}> $')
ax[0].legend()
ax[0].grid(True)

x, y = np.loadtxt("Datas/path_square.dat", usecols=(0,1), unpack='true')
ax[1].plot(x, y, marker = "o")
ax[1].set_title("Best path (square)")
ax[1].set(xlabel="x", ylabel="y")
ax[1].grid(True)

plt.subplots_adjust(left=0.2, bottom=0., right=2.2, top=1.2, wspace=0.3, hspace=0.4) 

plt.show()
```

Again, the same observations made above can be applied to this case, i.e $\langle L^{(1)}\rangle$ averaged on the best half of the population tends to the minimum value as the number of generations increases.
