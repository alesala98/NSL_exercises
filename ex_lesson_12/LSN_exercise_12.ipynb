## <span style="color:red"> Exercise 12.1. </span>

The goal of exercise lesson 12 is to use and make practice with deep neural network models in order to recognize and distinguish ten handwritten digits (0-9).

In particular, by keeping fixed all the other parameters, using three other optimizers different from SGD, the deep neural network was trained on MNIST dataset, paying careful attention to accuracy and loss for training and validation data. Below, you can find the given script slightly modified in order to use different optimizers.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.layers import Dense as Dense
from tensorflow.keras.layers import Dropout as Dropout
from tensorflow.keras.layers import Flatten as Flatten
from tensorflow.keras.layers import Conv2D as Conv2D
from tensorflow.keras.layers import MaxPooling2D as MaxPooling2D
from tensorflow.keras.datasets import mnist
```

```python
np.random.seed(0)
tf.random.set_seed(0)

img_rows, img_cols = 28, 28   # Input image dimensions: number of pixels 
num_classes = 10              # Output: 10 digits

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# Reshape data
X_train = X_train.reshape(X_train.shape[0], img_rows * img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows * img_cols)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

Y_train = tf.keras.utils.to_categorical(Y_train, num_classes)
Y_test = tf.keras.utils.to_categorical(Y_test, num_classes)
```

```python
def create_DNN():
    # Instantiate model
    model = tf.keras.Sequential()
    # Add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # Add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # Apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # Soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

def compile_model(optimizer):
    # Create the model
    model = create_DNN()
    # Compile the model
    model.compile(loss=tf.keras.losses.categorical_crossentropy, optimizer=optimizer, metrics=['accuracy'])
    
    return model
```

```python
# Training parameters
batch_size = 32
epochs = 5
Optimizers = ["sgd","RMSprop","Adam","Nadam"]

i = 1

for optimizer in Optimizers:
    
    # Create the deep neural net
    model_DNN = compile_model(optimizer)

    # Train DNN and store training info in history
    history = model_DNN.fit(X_train, Y_train, batch_size = batch_size, epochs=epochs,
                            verbose = 0, validation_data = (X_test, Y_test))
    
    # Evaluate model
    score = model_DNN.evaluate(X_test, Y_test, verbose = False)

    # Print performances
    print('Optimizer: ', optimizer)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    
    plt.figure(i, figsize = (18, 7))

    # Summarize history for accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.ylabel('Model accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc = 'best')
    plt.grid(True)

    # Summarize history for loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.ylabel('Model loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='best')
    plt.grid(True)
    
    plt.show()
    
    i += 1
```

Looking at all the plots above, first of all we can see that our neural network is characterised by very high accuracy rates. In fact, for all the optimizers, the accuracy rate is higher than 95%. 

Then, it seems that both Adam and Nadam optimizers give back the best results. However, it shuld be noted that, at some epoch-point, these two optimizers may be overfitting the training datas. Indeed, if you take a look at the loss functions trends, you can clearly see that while training curves continue to decrease, test curves remain more or less constant. This means that increasing the training time, which means increasing the number of epochs, could lead our neural network to obtain poorer results. On the other hand, SGD and RMSprop optimizers seem to not overfit training datas. Therefore, for these two models it should be safe to increase the training time to obtain better results.


## <span style="color:red"> Exercises 12.2.</span>

Then, the previously implemented deep neural network is modified adding the possibility to work using convolutional layers. This is exactly implemented in the second cell below, while the rest of the code is the same as the previous one. Again it performs an extimation of the performances of the NN. 

```python
img_rows, img_cols = 28, 28   # Input image dimensions: number of pixels 
num_classes = 10              # Output: 10 digits

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

if tf.keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

Y_train = tf.keras.utils.to_categorical(Y_train, num_classes)
Y_test = tf.keras.utils.to_categorical(Y_test, num_classes)
```

```python
def create_CNN():
   
    model = tf.keras.Sequential()
    model.add(Conv2D(20, kernel_size=(5, 5), activation='relu', input_shape=input_shape))                           
    model.add(MaxPooling2D(pool_size = (2,2)))
    model.add(Conv2D(10, kernel_size=(2, 2), activation='relu'))
    model.add(Dropout(rate=0.5))
    model.add(Flatten())
    model.add(Dense(50, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    model.compile(loss=tf.keras.losses.categorical_crossentropy, optimizer='SGD', metrics=['accuracy'])
    
    return model
```

```python
from PIL import Image
import os

# Training parameters
batch_size = 32
epochs = 5

# Create the deep convolutional network
model_CNN=create_CNN()

# Train CNN
history = model_CNN.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, 
                        validation_data=(X_test, Y_test), verbose = False)

# Evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose = False)

# Print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

plt.figure(figsize = (18, 8))

# Summarize history for accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('Model accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)

# Summarize history for loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('Model loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)

plt.show()
```

Finally, using gimp application ten pictures of all the digits between 0 and 9 were produced in order to prove if the CNN previously created and trained is able to recognize them. The results are shown below.

```python
plt.figure(figsize=(20,5))
print()

for k in range(10):
    digit_filename = "./images/"+str(k)+".png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    data /= 255
    
    plt.subplot(1, 10, k+1)   
    plt.imshow(data, cmap='gray')
    
    data = data.reshape(1,xdim*ydim)

    if tf.keras.backend.image_data_format() == 'channels_first':
        data = data.reshape(data.shape[0], 1, img_rows, img_cols)
    else:
        data = data.reshape(data.shape[0], img_rows, img_cols, 1)
        
    pred_0 = model_CNN.predict(data)
       
    plt.title("Digit predicted: {}".format(np.argmax(pred_0)))
    plt.axis('off')

plt.show()
```


As you can see, the CNN is able to recognize 8 digits out of ten. In particular, it isn't able to distinguish 7 from 2 and 9 from 8. Probably, this is caused by the fact that these digits are quite similar to each other from the point of view of the NN, while all the others are characterized by different shapes.
